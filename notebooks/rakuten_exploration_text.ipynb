<h3><font size=5> 1.1. Problématique <a name="Problématique"></a></h3>

Dans le cadre de ce projet, nous nous intéressons à la classification multimodale à grande échelle des produits sur la plateforme Rakuten France, en utilisant à la fois des données textuelles et visuelles. La problématique centrale réside dans la nécessité de catégoriser efficacement un vaste catalogue de produits, qui comprend des titres, des descriptions et des images, en codes de type de produit spécifiques. Cette tâche est d'une importance cruciale pour améliorer la recherche personnalisée et les recommandations, tout en faisant face aux défis liés à la diversité des produits, à la qualité variable des données et à la distribution déséquilibrée des classes. L'objectif est de développer un modèle de classification capable de traiter ces informations hétérogènes afin de réduire la duplication et d'optimiser l'expérience utilisateur sur la plateforme.

<h3><font size=5> 1.2. Présentation des données <a name="PrezDonnees"></a></h3>

Pour ce projet, nous disposons d'un ensemble de données comprenant environ 99 000 annonces de produits au format CSV, qui se divise en deux ensembles : un ensemble d'entraînement et un ensemble de test. Voici les détails des données :

<h4><font size=4> 1.2.1. Le jeu de données textuel <a name="Donnees_Textuelles"></a></h4>

**Ensemble d'Entraînement (X_train.csv et Y_train.csv) :**
   - **X_train.csv** : Contient les informations d'entrée pour l'entraînement, avec les colonnes suivantes :
     - **ID** : Un identifiant entier pour chaque produit.
     - **designation** : Le titre du produit, un texte court résumant le produit.
     - **description** : Un texte plus détaillé décrivant le produit (peut contenir des valeurs manquantes).
     - **productid** : Un identifiant unique pour le produit.
     - **imageid** : Un identifiant unique pour l'image associée au produit.
   - **Y_train.csv** : Contient les sorties d'entraînement, avec la colonne :
     - **prdtypecode** : Le code de type de produit correspondant à chaque ID dans X_train.csv.  

**Ensemble de Test (X_test.csv) :**
   - **X_test.csv** : Contient les informations d'entrée pour le test, avec la même structure que X_train.csv. Les participants doivent prédire les codes de type de produit pour cet ensemble.

```python
# Calculer le nombre total de produits pour chaque type de produit
nombre_total_produits = merged_df.groupby('prdtypecode')['description'].count()

# Calculer le nombre de descriptions manquantes pour chaque type de produit
nombre_manquant = merged_df.groupby('prdtypecode')['description'].apply(lambda x: x.isnull().sum())

# Calculer le pourcentage de descriptions manquantes
pourcentage_manquant = (nombre_manquant / (nombre_total_produits + nombre_manquant)) * 100

# Convertir en DataFrame
resultat_df = pd.DataFrame({
    'Type de Produit': nombre_total_produits.index,
        'Pourcentage Manquant (%)': pourcentage_manquant.round(2),
    'Nombre Total de Produits (non null)': nombre_total_produits.values,
})

# Réinitialiser l'index
resultat_df = resultat_df.reset_index(drop=True)

# Convertir type_counts en DataFrame si ce n'est pas déjà fait
type_counts_df = type_counts.reset_index()
type_counts_df.columns = ['Type de Produit', 'Nombre de Produits']  # Renommer les colonnes

# Fusionner avec type_counts_df
resultat_df = pd.merge(resultat_df, type_counts_df, on='Type de Produit', how='left')

# Trier le DataFrame selon le pourcentage manquant (du plus grand au plus petit)
resultat_df = resultat_df.sort_values(by='Pourcentage Manquant (%)', ascending=False)

# Appliquer un style au DataFrame
styled_df = resultat_df.style \
    .set_table_attributes('style="width: 100%; border-collapse: collapse;"') \
    .set_caption("Analyse des Produits par Type") \
    .set_properties(**{'text-align': 'left', 'font-size': '14px'}) \
    .format({'Pourcentage Manquant (%)': '{:.2f}%'}) \
    .background_gradient(subset=['Pourcentage Manquant (%)'], cmap='viridis') \
    .background_gradient(subset=['Nombre de Produits'], cmap='viridis') \
    .background_gradient(subset=['Nombre Total de Produits (non null)'], cmap='viridis') \
    .applymap(lambda x: 'background-color: lightcoral' if pd.isnull(x) else '', subset=['Nombre Total de Produits (non null)'])

# Afficher le DataFrame stylisé
styled_df
```

Que la description soit manquante ou non, nous indique clairement quelque chose. Il serait judicieux de ne pas supprimer ces données par la suite. Une idée serait de rendre cette information binaire : 0 si il n'y pas de description, 1 sinon.

```python
# Fonction pour classer les descriptions
def classer_description(description):
    if pd.isna(description) or description.strip() == "":
        return 'Vide'
    elif re.search(r'<[^>]+>', description):  # Vérifie si la description contient du HTML
        return 'HTML'
    elif re.search(r'^\s*#', description):  # Vérifie si la description commence par un # (Markdown)
        return 'Markdown'
    elif description.strip().startswith('{') and description.strip().endswith('}'):  # Vérifie si c'est du JSON
        return 'JSON'
    elif re.search(r'<\?xml', description):  # Vérifie si la description commence par une déclaration XML
        return 'XML'
    elif re.search(r'\\begin\{.*\}', description):  # Vérifie si la description contient des commandes LaTeX
        return 'LaTeX'
    elif re.search(r'^\s*---', description):  # Vérifie si la description commence par des tirets (YAML)
        return 'YAML'
    else:
        return 'Autre'

# Appliquer la fonction à la colonne de description
merged_df['classe_description'] = merged_df['description'].apply(classer_description)
merged_df['classe_designation'] = merged_df['designation'].apply(classer_description)

# Afficher le résultat
type_counts_description = merged_df['classe_description'].value_counts()
type_counts_designation = merged_df['classe_designation'].value_counts()

print(f"Répartition des formats de description : \n{type_counts_description}")
print(f"\n\nRépartition des formats de designation : \n{type_counts_designation}")
```

```python
# Liste des mots de liaison à exclure
mots_vides = set([
    'de', 'le', 'la', 'et', 'à', 'dans', 'un', 'une', 'pour', 'en', 
    'que', 'qui', 'sur', 'avec', 'par', 'ce', 'cet', 'cette', 'il', 
    'elle', 'nous', 'vous', 'ils', 'elles', 'au', 'aux', 'des', 
    'est', 'sont', 'a', 'avoir', 'être', 'comme', 'mais', 'ou', 
    'si', 'non', 'tout', 'tous', 'toute', 'toutes', 'bien', 'plus',
    'les', 'votre', 'couleur', 'taille', 'x', 'br', 'du', 'pas', 'cm',
    '1', 'taille', 'peut', 'haute', 'ne', 'qualité', '2', 'l', 'p',
    'facile', '100', '3', 'and', 'sans', '4', 'dimensions', 'm', 'se',
    'plaît', 'vos', 'the', 'nos', 'ces', 'ni', 'bbr', 'son', 'sa', 'très',
    'leur', '43', '10', 'pbr', 'pp', '3939', 'of', 'h',
])

# Fonction pour nettoyer et extraire les mots
def extraire_mots(description):
    # Nettoyer la description : enlever la ponctuation et mettre en minuscules
    description = re.sub(r'[^\w\s]', '', description)
    mots = description.lower().split()
    # Filtrer les mots vides
    return [mot for mot in mots if mot not in mots_vides]

# Dictionnaire pour stocker les mots fréquents par classe
mots_frequents_par_classe = {}

# Regrouper par classe de produit
for produit, groupe in merged_df.groupby('prdtypecode'):
    tous_les_mots = []
    for description in groupe['description'].dropna():  # Ignorer les valeurs NaN
        tous_les_mots.extend(extraire_mots(description))
    
    # Compter les occurrences des mots
    compteur = Counter(tous_les_mots)
    
    # Obtenir les 10 mots les plus fréquents
    mots_frequents = compteur.most_common(10)
    
    # Stocker les résultats dans le dictionnaire
    mots_frequents_par_classe[produit] = mots_frequents

# Afficher les résultats
for produit, mots in mots_frequents_par_classe.items():
    print(f"\nClasse de produit : {produit}")
    for mot, freq in mots:
        print(f"{mot}: {freq} fois")

# Optionnel : sauvegarder les résultats dans un fichier CSV
resultats = []
for produit, mots in mots_frequents_par_classe.items():
    for mot, freq in mots:
        resultats.append({'Produit': produit, 'Mot': mot, 'Fréquence': freq})

resultats_df = pd.DataFrame(resultats)
resultats_df.to_csv('mots_frequents_par_classe.csv', index=False)
```
